### Dog Ratings

WeRateDogs is a Twitter account that is involved in the professional rating of people's dogs with a humorous comment about the dog. They have about 9.3 millions followers on twitter and a unique ratings system in which the denominator almost always have a value of 10 while The numerators are almost always greater than 10.
An Assessment and Analysis of about 959 tweets from @dog_rates twitter handle generated between the 15th of November 2015 and 1st of August 2017 involving about one hundred and eleven breeds of dogs at various growth stages revealed some of these findings;

- Golden retriever dog breeds appears to be the most commonly rated dogs by WeRAteDogs within the time frame in which the tweets were generated with about 89 dogs followed by Pembroke 60, Labrador_retriever 58 and Chihuahua 52 while The most commonly observed growth stage of these rated dogs were the Pupper (113) followed by doggo (38), puppo (18) and floofer (6) out of the 175 tweets in which the dog stages where available for extraction.

- Also, analysis showed that The Standard poodle, Mexican hairless and English springer dog breeds seeems to have the highest mean retweet counts while the Mexican_hairless, Black-and-tan_coonhound and Standard_poodle have the highest mean favorite counts in descending order.
